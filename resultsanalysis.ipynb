```python
import json

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
sns.set_context('talk')
```

```python
with open('./results_2-1.json', 'r') as f:
    results1 = json.load(f)

with open('./results_2-2.json', 'r') as f:
    results2 = json.load(f)

with open('./results_2-3.json', 'r') as f:
    results3 = json.load(f)

with open('./results_2-4.json', 'r') as f:
    results4 = json.load(f)
```

```python
results = dict()
# results['train_loss'] = 
```

```python
results['train_loss'] = results1['train_loss'] + results2['train_loss'] + results3['train_loss'] + results4['train_loss']
results['train_correlation'] = results1['train_correlation'] + results2['train_correlation'] + results3['train_correlation'] + results4['train_correlation']
results['train_r2'] = results1['train_r2'] + results2['train_r2'] + results3['train_r2'] + results4['train_r2']
results['validation_loss'] = results1['validation_loss'] + results2['validation_loss'] + results3['validation_loss'] + results4['validation_loss']
results['validation_correlation'] = results1['validation_correlation'] + results2['validation_correlation'] + results3['validation_correlation'] + results4['validation_correlation']
results['validation_r2'] = results1['validation_r2'] + results2['validation_r2'] + results3['validation_r2'] + results4['validation_r2']
```

```python
print('Len results 1', len(results1['train_loss']))
print('Len results 2', len(results2['train_loss']))
print('Len results 3', len(results3['train_loss']))
print('Len results 4', len(results4['train_loss']))
```

```python
results_df = pd.DataFrame.from_records(results)
```

```python
# high dpi retina display
%config InlineBackend.figure_format = 'retina'
```

```python
plt.figure(figsize=(18,5))
plt.subplot(1,3,1)

plt.plot(results_df['train_loss'], label='Train')
plt.plot(results_df['validation_loss'], label='Val')
# plt.yscale('log')
plt.xlabel('Epoch')
# vertical lines at 400, 450, 550, and 950 epochs
plt.axvline(400, color='k', linestyle='--', alpha=0.3)
plt.axvline(450, color='k', linestyle='--', alpha=0.3)
plt.axvline(550, color='k', linestyle='--', alpha=0.3)
# plt.axvline(950, color='k', linestyle='--', alpha=0.3)
plt.title('Loss')
plt.legend()

plt.subplot(1,3,2)
plt.plot(results_df['train_correlation'], )#label='Train corr')
plt.plot(results_df['validation_correlation'], )#label='Val corr')
# plt.ylim(-1,1)
# plt.yscale('log')
plt.xlabel('Epoch')
plt.axvline(400, color='k', linestyle='--', alpha=0.3)
plt.axvline(450, color='k', linestyle='--', alpha=0.3)
plt.axvline(550, color='k', linestyle='--', alpha=0.3)
# plt.axvline(950, color='k', linestyle='--', alpha=0.3)
plt.title('Correlation')
# plt.legend()

plt.subplot(1,3,3)
plt.plot(results_df['train_r2'], )#label='Train R2')
plt.plot(results_df['validation_r2'], )#label='Val R2')
# plt.ylim(-1,1)
# plt.yscale('log')
plt.xlabel('Epoch')
plt.axvline(400, color='k', linestyle='--', alpha=0.3)
plt.axvline(450, color='k', linestyle='--', alpha=0.3)
plt.axvline(550, color='k', linestyle='--', alpha=0.3)
# plt.axvline(950, color='k', linestyle='--', alpha=0.3)
plt.title('R2')
# plt.legend()
```
